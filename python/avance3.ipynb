# Graphs

```python
fig = px.bar(x = ["Domingo","Lunes","Martes","Miércoles","Jueves","Viernes","Sábado"],y= accidents["day_of_week"].value_counts()[::-1],
            width=800, height=400,
            labels = dict(x = "Días",y="Cantidad de accidentes"))
fig.show()
```

```python
hora = [i//3 for i in range(24*3)]
gravedad = [ ["Grave","Moderado","Leve"][::-1][int(i%3)] for i in range(24*3)]
cantidad = [len(accidents.query(f"time == {hora[i]} and accident_severity == {3-int(i%3)}")) for i in range(24*3)]
df_aux = pd.DataFrame(data = {"Hora":hora,
                              "Gravedad":gravedad,
                              "Cantidad de accidentes":cantidad
                              })

draft_template = go.layout.Template()
draft_template.layout.annotations = [
    dict(
        name="Team",
        text="Team Van Rossum",
        textangle=0,
        opacity=0.1,
        font=dict(color="black", size=15),
        xref="paper",
        yref="paper",
        x=0,
        y=1.1,
        showarrow=False,
    )]

fig = px.bar(df_aux, x="Hora", y="Cantidad de accidentes", color="Gravedad",
            width=800,height=400,
            color_discrete_map={
                'Leve': '#99cc33',
                'Moderado': '#ffcc00',
                "Grave":"#cc3300"},
            template=draft_template)
del df_aux
fig.show()
#fig.write_image("hora.svg", engine="kaleido")
```
